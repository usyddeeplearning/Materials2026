## 3. PyTorch Tensor Operations

**Tensors** are the fundamental data structure in PyTorch — multi-dimensional arrays analogous to NumPy arrays, but with two critical additions: **GPU acceleration** and **automatic differentiation** (needed for backpropagation). In deep learning, all inputs, weights, gradients, and intermediate activations are represented as tensors.

The examples below show PyTorch and NumPy side-by-side, so you can see the correspondence between the two APIs and understand when to use each.

**Create a 5×5 integer matrix** with random values in $[0, 5)$:

````python
# by pytorch
import torch

# torch.randint(lower_bound, upper_bound, (rows, columns))
A = torch.randint(0, 5, (5, 5))

print(A)
````

````python
# by numpy
import numpy as np

# numpy.random.randint(lower_bound, upper_bound, (rows, columns))
A_np = np.random.randint(0, 5, (5, 5))

print(A_np)
````

````python
A[3][2]
````

### 3.1 Basic Tensor Operations

For matrices $A \in \mathbb{R}^{m \times n}$ and $B \in \mathbb{R}^{m \times n}$, the following operations are commonly used in deep learning (e.g., in computing weighted sums, loss functions, and gradient updates):

| Operation | Math notation | PyTorch | NumPy |
|-----------|--------------|---------|-------|
| Addition | $C = A + B$ | `A + B` | `A + B` |
| Subtraction | $C = A - B$ | `A - B` | `A - B` |
| Element-wise product | $C_{ij} = A_{ij} \cdot B_{ij}$ | `A * B` | `A * B` |
| Matrix product | $C = A B^T$ | `torch.matmul(A, B.T)` or `A @ B.T` | `np.matmul(A, B.T)` |

Define $A \in \mathbb{R}^{5 \times 6}$, $B \in \mathbb{R}^{5 \times 6}$:

````python
A = torch.randint(0, 5, (5, 6))
print(A)

B = torch.randint(0, 5, (5, 6))
print(B)
````

````python
A_np = np.random.randint(0, 5, (5, 6))
print(A_np)

B_np = np.random.randint(0, 5, (5, 6))
print(B_np)
````

* **Add**: $C = A + B$
* **Substract**: $C = A - B$

````python
# implement A + B in torch
C = A + B
print(C)

# implement A + B in numpy
C_np = A_np + B_np
print(C_np)

# implement A - B in torch
C = A - B
print(C)

# implement A - B in numpy
C_np = A_np - B_np
print(C_np)
````

* **Element-wise Multiplication** $C = A \odot B \in \mathbf{R}^{m \times n}, \forall i,j$, which means $C_{i, j} = A_{i, j} * B_{i, j}$.

````python
# implement A * B in torch
C = A * B
print(C)

# implement A * B in numpy
C_np = A_np * B_np
print(C_np)
````

* **Multiplication** $C = A \times B^T$, which means $C_{i,j} = ∑_{l=1}^{n}A_{i,l} * B_{l, j}$

* $A_{m\times n}, B_{n \times k}$ so we have output $C_{m \times k}$

````python
# implement A * B in torch
C = torch.matmul(A, B.T)
print(C)

# implement A * B in numpy
C_np = np.matmul(A_np, np.transpose(B_np))
print(C_np)
````

````python
# some other ways to implement A * B in torch
C = A @ B.T
print(C)

C = A.mm(B.T)
print(C)
````

### 5.1 Introduction

**Claude Code** is Anthropic's official AI-powered coding assistant that runs directly in your terminal or IDE. Unlike browser-based chatbots, Claude Code has **direct access to your local files** — it can read, write, edit, and reason about your entire project without you copying and pasting anything.

**Why is this relevant for a deep learning course?**

| Task | How Claude Code helps |
|------|-----------------------|
| Understanding unfamiliar PyTorch APIs | Ask in plain English; Claude reads your actual code to answer |
| Debugging shape mismatches and errors | Paste the traceback; Claude diagnoses the root cause in context |

| Explaining paper implementations | Ask what a specific code block does mathematically |

---

> ### CRITICAL: Always Review AI-Generated Code
>
> **Treating Claude Code output as correct without review is dangerous.** AI assistants can produce code that looks convincing but contains subtle bugs, incorrect assumptions, or logic errors — especially in deep learning, where mistakes in loss functions, gradient flow, or data preprocessing are often silent (no crash, but wrong results).
>
> **Specific risks in this course:**
> - **Incorrect tensor shapes** — generated code may silently broadcast where it should fail.
> - **Wrong loss or metric** — a plausible-looking function may compute the wrong quantity for your task.
> - **Training loop errors** — missing `.zero_grad()`, double-counting samples, or incorrect averaging can corrupt training silently.
> - **Data leakage** — normalisation or augmentation applied before splitting can leak test information into training.
>
> **The right mindset:** Use Claude Code as a fast first draft, not a final answer. You are responsible for understanding and verifying every line of code that runs in your experiments.
>
> - Read the generated code line by line before running it.
> - Test on small inputs with known expected outputs.
> - Cross-check against lecture slides, documentation, or reference implementations.
> - If you cannot explain what a generated line does, do not accept it.

#### Permission Modes

Claude Code has three modes that control how much autonomy it has over your files:

| Mode | What it does | When to use |
|------|-------------|-------------|
| **Auto** | Applies all file edits automatically without asking for approval each time | Rapid prototyping — you trust Claude to modify files freely within the session |

> **Note — file creation and shell commands:** These modes govern edits to *existing* files. Regardless of mode, Claude Code will still prompt for confirmation before **creating new files or executing shell commands** (which could delete or overwrite files). To bypass all permission checks — for example, in an automated pipeline or a sandboxed container — start Claude Code with:
> ```
> claude --dangerously-skip-permissions
> ```
> Only use this flag in isolated environments (e.g. a Docker container or VM) where unintended changes cannot cause lasting harm.
